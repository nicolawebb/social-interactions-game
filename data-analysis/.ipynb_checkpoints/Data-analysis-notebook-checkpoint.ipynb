### Prepare data

The dataframe `df` only contains the pose of the player and NPCs, transformed from Godot (Y up) frame to the more common maths/robotics frame (Z up). Columns are also renamed with shorter names.

```python
df = raw_df[["p_id"]].copy()

# creates a frame index for each participant, by substracting the first timestamp of each participant to the current timestamp
# df["frame"] = raw_df.groupby("p_id")["ticks_msec"].transform(lambda x: round((x - x.min())/1000.))



for col in raw_df.columns:
    
    npcname = re.sub(r'ProximityGroup._Townsperson', 'n', col)
    npcname = re.sub(r'player', 'p', npcname)
    npcname = re.sub(r'ProximityGroup._Detective', 'd', npcname)
    
    if "_x" in col:
        df[npcname[:-1] + "x"] =  raw_df[col]
        df[npcname[:-1] + "z"] = 2.157
    elif "_z" in col:
        df[npcname[:-1] + "y"] =  -raw_df[col]
        
    elif 'rx' in col:
        df[npcname[:-2] + "roll"] =  0
    elif 'ry' in col:
        df[npcname[:-2] + "pitch"] =  0
    elif 'rz' in col:
        df[npcname[:-2] + "yaw"] =  np.radians(raw_df[col])-np.pi/2

# use the index as frame index
df = df.reset_index().rename(columns={"index":"frame"})

df
```

## Angle/FOV

First, compute the position of the NPCs in the player's frame of reference, and the position of the player in each of the NPC's frames of reference:

```python


def get_transform_world_2_player(array, agls):
    Tpw = t.compose_matrix(translate=[array[0], array[1], array[2]], angles=agls) #returns 4x4
    Twp = np.linalg.inv(Tpw)
    return(Twp)

def transform_npc(px, py, pz, p_rx, p_ry, p_rz, nx, ny, nz):
    Twp = get_transform_world_2_player([px, py, pz], [p_rx, p_ry, p_rz])
    return Twp.dot(np.array([float(nx),float(ny),float(nz),1.0]))

for idx in range(1,10):
    npc_x = 'n%s_x'%idx
    npc_y = 'n%s_y'%idx
    npc_z = 'n%s_z'%idx
    npc_pitch = 'n%s_pitch'%idx
    npc_roll = 'n%s_roll'%idx
    npc_yaw = 'n%s_yaw'%idx

    df["pn%sx"%idx] = df.apply(lambda row: transform_npc(row[npc_x], row[npc_y], row[npc_z],row[npc_pitch], row[npc_roll], row[npc_yaw], row['p_x'], row['p_y'], row['p_z'])[0], axis=1)
    df["pn%sy"%idx] = df.apply(lambda row: transform_npc(row[npc_x], row[npc_y], row[npc_z],row[npc_pitch], row[npc_roll], row[npc_yaw], row['p_x'], row['p_y'], row['p_z'])[1], axis=1)
    df["pn%sz"%idx] = df.apply(lambda row: transform_npc(row[npc_x], row[npc_y], row[npc_z],row[npc_pitch], row[npc_roll], row[npc_yaw], row['p_x'], row['p_y'], row['p_z'])[2], axis=1)

    df["n%spx"%idx] = df.apply(lambda row: transform_npc(row['p_x'], row['p_y'], row['p_z'],row['p_pitch'], row['p_roll'], row['p_yaw'], row[npc_x], row[npc_y], row[npc_z])[0], axis=1)
    df["n%spy"%idx] = df.apply(lambda row: transform_npc(row['p_x'], row['p_y'], row['p_z'],row['p_pitch'], row['p_roll'], row['p_yaw'], row[npc_x], row[npc_y], row[npc_z])[1], axis=1)
    df["n%spz"%idx] = df.apply(lambda row: transform_npc(row['p_x'], row['p_y'], row['p_z'],row['p_pitch'], row['p_roll'], row['p_yaw'], row[npc_x], row[npc_y], row[npc_z])[2], axis=1)
    
df
```

# Metric of Social Engagement

This metric is calculated as the product of the mutual optical axis distance of two persons, divided by the distance between these two persons.

```python
def fov_dist(df, id):
    """Calculate how 'far' from the optical axis (X-axis) a given point is.
    The value if normalised from 1 (on the optical axis) to 0 (at the edge of the field of view)
    """
    series = (1-abs(df['%sy'%id]/(tan*df['%sx'%id]))).clip(0)
    
    # if the other person is *behind*, set the 'fov_dist' to zero (not visible)
    series[df['%sx'%id] < 0] = 0
    return series


for i in range(1,10):
    df["fov_pn%s" % i] = fov_dist(df, "pn%s" % i)
    df["fov_n%sp" % i] = fov_dist(df, "n%sp" % i)
    
    df["fov_%s" % i] = (df["fov_n%sp" % i] * df["fov_pn%s" % i])
    
    df["social_engagement_%s" % i] = (df["fov_%s" % i] / df["dist_n%s" % i]).clip(0,1)
    
```

## Interaction profiles

A set window of time is extracted around each of the interactions with the NPCs (defined as when the player enters the close-proximity zone, which triggers the NPCs dialogue). For each participant, we average the 'social engagement' metric over that window of interaction, to create 'interaction profiles'.

```python
# 15 seconds after and before the start of the interaction
WINDOW_SIZE=15
```

```python
interactions = {"p_id": [], "frame": [], "social_engagement": []}
interaction_id = 0  

interactions_count = {"p_id": [], "count": []}


for p_id in df["p_id"].unique():
    
    p_df = df[df["p_id"] == p_id]
    count = 0  
    
    for idx in range(1,10):
        npc = "dist_n%s" % idx

        # this expression selects the dataframe rows corresponding to 'just' entering into the 1.2m interaction zone.
        selector = ((p_df[npc] < 1.2) ^ (p_df[npc] < 1.2).shift(1,fill_value=False)) & (p_df[npc] < 1.2)

        last_index = -WINDOW_SIZE
        

        
        for index, row in p_df[selector].iterrows():
            if index < last_index + WINDOW_SIZE:
                continue
            last_index = index
            count += 1

            for window_offset in range(-WINDOW_SIZE,WINDOW_SIZE):
                frame_idx = row["frame"] + window_offset

                if len(p_df[p_df["frame"] == frame_idx]):
                    interactions["p_id"].append(p_id)
                    interactions["frame"].append(window_offset)
                    if len(p_df[p_df["frame"] == frame_idx]) > 1:
                        display(p_df[p_df["frame"] == frame_idx])
                    interactions["social_engagement"].append(float(p_df[p_df["frame"] == frame_idx]["social_engagement_%s" % idx]))
        
    interactions_count["p_id"].append(p_id)
    interactions_count["count"].append(count)

interaction_df = pd.DataFrame(interactions)    
interactions_count_df = pd.DataFrame(interactions_count)

# pd.set_option("max_columns", None)
# pd.set_option("max_rows", None)
interaction_df[0:30]
```

## Correlations between social engagement profiles and personality

Features extraction:

```python
# only keep participants whose demographics we have
interaction_profiles = interaction_profiles[interaction_profiles["p_id"].isin(dfbig5.reset_index()["p_id"])]

soceng_mean = interaction_profiles.groupby("p_id")["social_engagement_mean"].mean().reset_index()
soceng_mean = soceng_mean.set_index("p_id")

soceng_std = interaction_profiles.groupby("p_id")["social_engagement_std"].mean().reset_index()
soceng_std = soceng_std.set_index("p_id")
```

```python

interaction_profiles_after = interaction_profiles[interaction_profiles["frame"]>0]
interaction_profiles_before = interaction_profiles[interaction_profiles["frame"]<0]

soceng_after_mean = interaction_profiles_after.groupby("p_id")["social_engagement_mean"].mean().reset_index()
soceng_after_mean = soceng_after_mean.set_index("p_id")
soceng_after_mean = soceng_after_mean.rename(columns={"social_engagement_mean":"social_engagement_after"})

soceng_before_mean = interaction_profiles_before.groupby("p_id")["social_engagement_mean"].mean().reset_index()
soceng_before_mean = soceng_before_mean.set_index("p_id")
soceng_before_mean = soceng_before_mean.rename(columns={"social_engagement_mean":"social_engagement_before"})

soceng_before_delta = soceng_after_mean.rename(columns={"social_engagement_after":"social_engagement_delta"})
soceng_before_delta["social_engagement_delta"] -= soceng_before_mean["social_engagement_before"]
```
